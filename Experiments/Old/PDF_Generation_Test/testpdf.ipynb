```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import matplotlib.pyplot as plt
import CustomModules.weasy_print as wp


df = pd.DataFrame({
    "month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "sales": [100, 140, 130, 170, 160, 210],
    "profit": [20, 35, 30, 45, 42, 60],
    "customers": [12, 18, 15, 22, 21, 29],
})


# Sales chart
fig_sales, ax = plt.subplots()
ax.plot(df["month"], df["sales"], marker="o")
ax.set_title("Sales by Month")
ax.set_xlabel("Month")
ax.set_ylabel("Sales")


# Profit chart
fig_profit, ax = plt.subplots()
ax.bar(df["month"], df["profit"])
ax.set_title("Profit by Month")
ax.set_xlabel("Month")
ax.set_ylabel("Profit")


blocks = [
    wp.header("Monthly Sales Report"),

    wp.text("This report summarizes monthly sales, profit, and customer activity."),

    wp.text(f"Rows: {len(df):,}"),
    wp.text(f"Columns: {len(df.columns):,}"),

    wp.header("Raw Data", level=2),
    wp.table(df, max_rows=20),

    wp.header("Summary Statistics", level=2),
    wp.table(df.describe().round(2)),

    wp.header("Charts", level=1),

    wp.group(
        wp.header("Sales by Month", level=2),
        wp.plot(fig_sales, alt="Sales by month", width="50%"),
        avoid_break_inside=True,
    ),

    wp.group(
        wp.header("Profit by Month", level=2),
        wp.plot(fig_profit, alt="Profit by month", width="50%"),
        avoid_break_inside=True,
    ),

    wp.header("Notes", level=1, new_page=True),
    wp.text("Sales and profit both increased over the observed period."),
]


extra_css = """
h1 {
    font-size: 24pt;
}

h2 {
    margin-top: 18px;
}

.data-table {
    font-size: 8pt;
}
"""


wp.build_pdf(
    blocks,
    "monthly_sales_report.pdf",
    title="Monthly Sales Report",
    header_text="Internal Analytics",
    footer=True,
    page_size="A4",
    margin="18mm",
    extra_css=extra_css,
)
```
